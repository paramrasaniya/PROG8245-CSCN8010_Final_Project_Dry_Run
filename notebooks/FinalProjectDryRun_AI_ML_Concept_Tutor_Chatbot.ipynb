## Section 2 – Load our concept knowledge base
In this step, we load the AI/ML concept file that our chatbot uses to answer questions.

```python
DATA_FILE = Path("../data/concept_knowledge.json")

with open(DATA_FILE, "r", encoding="utf-8") as f:
    knowledge_base = json.load(f)

print("Number of concepts loaded:", len(knowledge_base["topics"]))
print("Topic names:")
for item in knowledge_base["topics"]:
    print("-", item["topic"])
```

### Talking Points
- In this cell, we loaded our concept knowledge base from the data folder.

- We also printed the topic names to verify that our data loaded correctly.

## Section 3 – Topic detection and multilingual answer function
Here we define the main function that identifies the concept and returns the answer in the selected language.

```python
LANG_MAP = {
    "English": "en",
    "Spanish": "es",
    "French": "fr",
    "Hindi": "hi",
    "Gujarati": "gu"
}

def detect_topic(question):
    if not question:
        return None
    text_lower = question.lower()
    best_match = None
    best_length = -1

    for item in knowledge_base["topics"]:
        if item["topic"] in text_lower and len(item["topic"]) > best_length:
            best_match = item
            best_length = len(item["topic"])
        for keyword in item["keywords"]:
            if keyword.lower() in text_lower and len(keyword) > best_length:
                best_match = item
                best_length = len(keyword)
    return best_match

def answer_question(question, answer_language="English"):
    lang = LANG_MAP.get(answer_language, "en")
    item = detect_topic(question)

    if item:
        return item["topic"], item["answers"].get(lang, item["answers"]["en"])

    fallback = "We could not match that exact topic. Please ask about our prepared AI/ML concepts."
    return "not found", fallback
```

### Talking Points
- In this cell, we created the core logic of our chatbot.
- First, we detect the topic from the user question using keyword matching.
- Then we return the answer in the language selected by the user.
- We also added a fallback message so the chatbot still responds even if the topic is not found.

## Section 4 – Test the chatbot with text questions
Now we test the chatbot using a few sample text questions.

```python
sample_questions = [
    "What is regression?",
    "Explain overfitting",
    "What is confusion matrix?",
    "What is a neural network?"
]

for q in sample_questions:
    topic, answer = answer_question(q, "English")
    print("Question:", q)
    print("Detected topic:", topic)
    print("Answer:", answer)
    print("-" * 60)
```

### Talking Points

- This helped us check if the topic detection and answer generation were working correctly.
- We used common AI/ML topics so our live demo questions can be predictable and clear.

## Section 5 – Test multilingual output
Here we show that the same topic can be answered in different languages.

```python
demo_question = "What is regression?"
for language in ["English", "Spanish", "French", "Hindi", "Gujarati"]:
    topic, answer = answer_question(demo_question, language)
    print(language + ":")
    print(answer)
    print()
```

### Talking Points

## Section 6 – Audio support functions
This section adds optional audio input and audio output support.

```python
SPEECH_LANG = {
    "English": "en-US",
    "Spanish": "es-ES",
    "French": "fr-FR",
    "Hindi": "hi-IN",
    "Gujarati": "gu-IN"
}

def transcribe_audio(audio_path, spoken_language="English"):
    if audio_path is None:
        return None, "No audio provided."
    if sr is None:
        return None, "SpeechRecognition is not installed."
    try:
        recognizer = sr.Recognizer()
        with sr.AudioFile(audio_path) as source:
            audio = recognizer.record(source)
        text = recognizer.recognize_google(audio, language=SPEECH_LANG.get(spoken_language, "en-US"))
        return text, f"Transcribed speech: {text}"
    except Exception as e:
        return None, f"Audio transcription failed: {e}"
```

### Talking Points
- In this cell, we added the audio transcription function.
- This gives our chatbot the audio-based feature by converting speech into text.
- We also handled errors carefully because microphone or speech setup can vary on different systems.
